```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('deeplearning.mplstyle')
```

```python
x_train = np.arange(1.0, 2.0, 0.125)
y_train = np.arange(300.0, 500.0, 25)
print(f"x_train={x_train}")
print(f"y_train={y_train}")
```

```python
print(f"x_train.shape={x_train.shape}")
m=x_train.shape[0]
print(f'number of training samples={m}')
```

```python
plt.scatter(x_train,y_train,marker='+',color='red')
plt.title('Housing Prices')
plt.xlabel('size(1000 sqft)')
plt.ylabel('price(in 1000s of dollars)')
plt.show()
```

```python
w=200
b=100
```

```python
def compute_model_output(x,w,b):
    m=len(x)
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=x[i]*w+b
    return f_wb   
```

```python
tmp_f_wb= compute_model_output(x_train,w,b)
plt.plot(x_train,tmp_f_wb,c='b',label='Our Prediction')
plt.scatter(x_train,y_train,marker='+',color='red',label='Training Data')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.plot()
```
